```python
#import and query a database using sqlite library

# import database file
import pandas.io.sql as pds
import pandas as pd
import sqlite3 as sq3

# initialise path
path = 'baseball.db'

# creating connection
con = sq3.Connection(path)
```

```python
# writing a query
query = '''SELECT * FROM schools;'''

# executing query
obs = pd.read_sql(query, con)
obs.head()
```

```python
# analyse allstarful table to rport the top players in terms off number of games played
# if there are ties, choose players with lower starting positions first

best_query = '''
    SELECT playerID AS numGamesPlayed,
    AVG(startingPos) AS avgStartingPosition
    FROM allstarfull
    GROUP BY playerID
    Order BY numGamesPlayed DESC, avgStartingPosition ASC
    LIMIT 3
'''
best = pd.read_sql(best_query, con)
# print(best.head())
display(best.head())
```
