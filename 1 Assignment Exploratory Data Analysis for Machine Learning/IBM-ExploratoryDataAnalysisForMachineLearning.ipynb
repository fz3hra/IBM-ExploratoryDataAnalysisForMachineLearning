# Brief description of Dataset and its attributes:

##### The UFO sightings Dataset contains over 80,000 records of UFO sightings since 1949 with latitude and longitude data allowing us to assess the global distribution of UFO sightings. The Date and time along with the duration of the UFO's stay and description of the craft also lend themselves to prediction. 


##### Content:
Date_time: standardized date and time of sighting

city: location of UFO sighting

state/province: the us state/canadian province, appears blank for other locations

Country: Country of UFO sighting

UFO_shape: a one word description of the "spacecraft"

length_of_encounter_seconds: standardized to seconds, length of the observation of the UFO

described_duration_of_counter: raw description of the length of the encounter (shows uncertainty to previous column)

Decription: text description of the UFO encounter. Warning column is messy, with some curation it could lend itself to some natural language processing and sentiment analysis

date_documented: when was the UFO sighting reported

latitude: latitude

longitude: longitude

```python
US = newData[newData.country=='us']
AU = newData[newData.country=='au']
```

```python
ttest_ind(US.yearOfSighting,AU.yearOfSighting, equal_var=False)
```

Interpretation:

if p<0.05, reject null hypothesis and accept alternate hypothesis.

if p>0.05 then fail to reject null and reject alternate.

p is 0.477 and more than 0.05, therefore we fail to reject the null.
